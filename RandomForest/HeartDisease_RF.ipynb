```python
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split 
from sklearn.ensemble import RandomForestClassifier 
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import roc_curve, auc
```

```python
df  = pd.read_csv("heart.csv")
```

```python
df
```

```python
# Cleaning data and making data ready
```

```python
#Lable Encoding for columns: 
#["Sex", "ChestPainType", "RestingECG", "ExerciseAngina", "ST_Slope"]
labelencoder = preprocessing.LabelEncoder()
categorical_cols = ["Sex", "ChestPainType", "RestingECG", "ExerciseAngina", "ST_Slope"]
for col in categorical_cols:
    df[col] = labelencoder.fit_transform(df[col])
```

```python
df
```

```python
#starting training process 
```

```python
#for the purpose of group work, all the group members used same traning and testing data with different methods
#From the original data, 70% was used for training and the rest (30%) for testing
```

```python
db_train = pd.read_csv("heartTraining.csv")
db_train
```

```python
db_test = pd.read_csv("heartTesting.csv")
db_test
```

```python
#getting the x and y values for training
```

```python
X_train = db_train.drop(columns="HeartDisease", axis=1)
X_train
```

```python
Y_train = db_train["HeartDisease"]
Y_train
```

```python
#getting the x and y values for testing 
```

```python
X_test = db_test.drop(columns="HeartDisease", axis=1)
X_test
```

```python
Y_test = db_test["HeartDisease"]
Y_test
```

```python
#starting the Random Forest Classification
```

```python
model = RandomForestClassifier()
model.fit(X_train, Y_train)
Y_predict = model.predict(X_test)
ac = accuracy_score(Y_test,Y_predict)
print('Accuracy is: ', ac*100)
```

```python
model = RandomForestClassifier(max_depth=5)
model.fit(X_train, Y_train)
Y_predict = model.predict(X_test)
ac = accuracy_score(Y_test,Y_predict)
print('Accuracy is: ', ac*100)
```

```python
model = RandomForestClassifier(max_depth=10)
model.fit(X_train, Y_train)
Y_predict = model.predict(X_test)
ac = accuracy_score(Y_test,Y_predict)
print('Accuracy is: ', ac*100)
```

```python
model = RandomForestClassifier(n_estimators=10)
model.fit(X_train, Y_train)
model.score(X_test, Y_test)
Y_predict = model.predict(X_test)
ac = accuracy_score(Y_test,Y_predict)
print('Accuracy is: ', ac*100)
```

```python
model = RandomForestClassifier(n_estimators=20)
model.fit(X_train, Y_train)
model.score(X_test, Y_test)
Y_predict = model.predict(X_test)
ac = accuracy_score(Y_test,Y_predict)
print('Accuracy is: ', ac*100)
```

```python
model = RandomForestClassifier(max_depth=10, n_estimators=10)
model.fit(X_train, Y_train)
model.score(X_test, Y_test)
Y_predict = model.predict(X_test)
ac = accuracy_score(Y_test,Y_predict)
print('Accuracy is: ', ac*100)
```

```python
model = RandomForestClassifier(max_depth=10, n_estimators=20)
model.fit(X_train, Y_train)
model.score(X_test, Y_test)
Y_predict = model.predict(X_test)
ac = accuracy_score(Y_test,Y_predict)
print('Accuracy is: ', ac*100)
```

```python
model = RandomForestClassifier(max_depth=5, n_estimators=30)
model.fit(X_train, Y_train)
model.score(X_test, Y_test)
Y_predict = model.predict(X_test)
ac = accuracy_score(Y_test,Y_predict)
print('Accuracy is: ', ac*100)
```

```python
model = RandomForestClassifier(max_depth=10, n_estimators=50)
model.fit(X_train, Y_train)
model.score(X_test, Y_test)
Y_predict = model.predict(X_test)
ac = accuracy_score(Y_test,Y_predict)
print('Accuracy is: ', ac*100)
```

```python
##########################################################
#                                                        #
# Best accuracy was found 86.95% with n_estimators=20    # 
#                                                        # 
##########################################################
```
